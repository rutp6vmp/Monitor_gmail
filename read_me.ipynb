寫成一個函式，可以將影像製作成MIMEImage格式的字串。

```python
from email.mime.image import MIMEImage
import cv2

import smtplib

def send_gmail(gmail_addr, gmail_pwd, to_addrs, msg):
    smtp_gmail = smtplib.SMTP('smtp.gmail.com', 587)# 建立 SMTP 物件
    print(smtp_gmail.ehlo())# Say Hello
    print(smtp_gmail.starttls())# 啟動 TLS 加密模式
    print(smtp_gmail.login(gmail_addr, gmail_pwd))# 登入
    status = smtp_gmail.sendmail(gmail_addr, to_addrs, msg)# 寄出
    if not status:
        print('寄信成功')
    else:
        print('寄信失敗', status)
    smtp_gmail.quit()# 結束與郵件伺服器的連線
def get_mime_img(subject, fr, to, img):
    #將 img 編碼為 JPGE 格式，[1]為返回資料, [0]為返回是否成功
    img_encode = cv2.imencode('.jpg', img)[1]
    # 再將資料轉為 bytes, 此即為要傳送的資料
    img_bytes = img_encode.tobytes()
    mime_img = MIMEImage(img_bytes)   # 建立 MIMEImage 物件
    mime_img['Content-type'] = 'application/octet-stream'
    mime_img['Content-Disposition'] = 'attachment; filename="pic.jpg"'
    mime_img['Subject'] = subject
    mime_img['From'] = fr
    mime_img['To'] = to
    return mime_img.as_string()   #←轉為字串後傳回

gmail_addr = 'gmail信箱'
gmail_pwd = 'gmail密碼'
to_addrs = ['回傳信箱1','回傳信箱2']

cap = cv2.VideoCapture(0)  # 打開攝影機
while cap.isOpened():
    success, img = cap.read()
    if success:
        cv2.imshow('frame', img)  # 在視窗中顯示影像
    k = cv2.waitKey(1)     # 讀取鍵盤輸入
    if k == ord('s'):   #←如果按下 s 鍵
        msg = get_mime_img('牛奶來襲', '牛奶防盜監視器', '壞人在偷吃腳腳', img)
        send_gmail(gmail_addr, gmail_pwd, to_addrs, msg) #以 gmail 寄出郵件
    elif k == ord('q'):   # 按下小寫 q 結束程式
        cv2.destroyAllWindows()
        cap.release()
        break
```

# OpenCV移動偵測
---
高斯模糊後，利用兩張影像相加減後取絕對值，判斷差異性。
    
diff = cv2.absdiff(此刻影像，前一刻影像)

```python
import cv2 

cap = cv2.cv2.VideoCapture(0)#建立攝影機影像
img_pre = None#前影像預設回None(空的)
while cap.isOpened():
    success, img = cap.read()#讀取此刻影像
    if success:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #影像灰階處理
        img_now = cv2.GaussianBlur(gray,(13,13),15)#高斯模糊
        if img_pre is not None:#如果前面的影像不是None就進行對比
            diff = cv2.absdiff(img_now , img_pre)#前後影像相減
            cv2.imshow('frame', diff)#顯示相減後的影像
        img_pre = img_now.copy()#再將現在的影像設為前影像
    
    k = cv2.waitKey(30)#稍作暫停並檢查按鍵
    if k == ord('q') :       #按下Q結束程式
        cv2.destroyAllWindows()
        cap.release()
        break
```

# 輪廓偵測:
---
我們現在有乾淨的影像了，接下來要對輪廓做偵測，相同顏色(像素值)的連續點，所聯成的一條曲線。

因為影像變化才會有像素值，而有像素質才能得到輪廓，進而判斷是否有物體經過。

```python
輪廓影像,輪廓座標,輪廓階層關係 = cv2.findContours(image,mode,method)
```

- image: 可以輸入灰階影像，為了準確性最好輸入只有0,255
- mode: 搜尋輪廓模式:
    * cv2.RETR_EXTERNAL :只搜尋最外層的輪廓
    * cv2.RETR_LIST:搜尋內外所有輪廓，但不建立階層關係
- method: 輪廓近似的表示法
    * cv2.CHAIN_APPROX_NONE:儲存所有輪廓點，兩個相鄰的點位置不超過1
    * cv2.CHAIN_APPROX_SIMPLE:降至1維，只留下方向的端點座標

若有偵測到輪廓會回傳一個三維的np.array(陣列),若沒有偵測到則回傳一個空的list，可以判斷影像是否出現變化。

---
####  繪製輪廓:

    偵測到輪廓後可以使用 `cv2.drawContours()`方法可以幫我們在指定影像上畫出輪廓:
    cv2.drawContours(image, contours, contourIdx, color, thickness)
    - image :輸入影像
    - contours: 輪廓座標的list
    - contourIdx: 可以指定要繪製哪個輪廓，若指定-1則繪製所有輪廓
    - color: 繪製的顏色，以tuple(B,G,R)指定
    - thickness: 繪製的寬度

# 程式碼說明:
- 1~11:進行前後影像相減
- 12 :相減後的影像，以門檻值25進行黑白化
- 14 搜尋黑白影像中的輪廓，若有找到在18.進行影像繪製

```python
import cv2

cap = cv2.cv2.VideoCapture(0)
img_pre = None   # 前影像, 預設是空的
while cap.isOpened():
    success, img = cap.read()
    if success:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # 灰階處理
        img_now = cv2.GaussianBlur(gray, (13, 13), 5)          # 高斯模糊
        if img_pre is not None:  #如果前影像不是空的, 就和前影像比對
            diff = cv2.absdiff(img_now, img_pre)     # 此影格與前影格的差異值
            ret, thresh = cv2.threshold(diff, 25, 255,  # 門檻值
                                        cv2.THRESH_BINARY)
            contours, _ = cv2.findContours(thresh,    # 找到輪廓
                                             cv2.RETR_EXTERNAL,
                                             cv2.CHAIN_APPROX_SIMPLE)
            if contours:    # 如果有偵測到輪廓
                cv2.drawContours(img, contours, -1, (255, 255, 255), 2)
                print('偵測到移動')
            else:
                print('靜止畫面')

        cv2.imshow('frame', img)
        img_pre = img_now.copy()
    k = cv2.waitKey(50)
    if k == ord('q'):
        cv2.destroyAllWindows()
        cap.release()
        break
```
